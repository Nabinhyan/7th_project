```python
import pandas as pd
from pandas import datetime
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def parser(x):
    return datetime.strptime(x,'%Y-%m-%d')

sales = pd.read_csv('importofrice.csv',index_col=0, parse_dates=[0] ,date_parser=parser)
```

```python
sales.head()
```

```python
sales.size
```

```python
sales.plot()
```

```python
plot_acf(sales)
```

```python
sales.shift(3)
```

```python
sales_diff = sales.diff(periods=1)
sales_diff=sales_diff[1:]
sales_diff.head()
```

```python
plot_acf(sales_diff)
```

```python
sales_diff.plot()
```

```python
X=sales.values
train=X[0:87]
test=X[87:]
predictions = []
```

```python
train.size
```

```python
test.size
```

```python
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error
model_ar = AR(train)
model_ar_fit=model_ar.fit()
```

```python
predictions=model_ar_fit.predict(start=96,end=120)
```

```python
test
```

```python
plt.plot(test)
plt.plot(predictions,color='red')
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model_arima=ARIMA(train,order=(27,0,1))
model_arima_fit=model_arima.fit()
print(model_arima_fit.aic)
```

```python
predictions=model_arima_fit.forecast(steps=24)[0]
predictions
```

```python
plt.plot(test)
plt.plot(predictions,color='green')
```

```python
mean_squared_error(test,predictions)
```
